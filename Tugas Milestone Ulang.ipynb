```python
import csv
import pandas as pd
data=pd.read_csv('DataMilestoneUlang.csv', encoding='utf-8')
```

```python
data
```

```python
saham=pd.read_csv('saham.csv',encoding='utf-8')
```

```python
DataMean=data.groupby(['MonthYear']).mean()[['AvgTone','GoldsteinScale','NumMentions']].reset_index()
```

```python
DataSum=data.groupby(['MonthYear']).sum()[['AvgTone','GoldsteinScale','NumMentions']].reset_index()
```

```python
DataCount=data.groupby(['MonthYear']).count()['AvgTone'].reset_index()
```

```python
MeanAvgTone=DataMean['AvgTone']
MeanGoldsteinScale=DataMean['GoldsteinScale']
MeanNumMentions=DataMean['NumMentions']
SumAvgTone=DataSum['AvgTone']
SumGoldsteinScale=DataSum['GoldsteinScale']
SumNumMentions=DataSum['NumMentions']
CountNews=DataCount['AvgTone']
NilaiSaham=saham['Nilai']
VolumeSaham=saham['Volume']
IndexHargaSaham=saham['IndexHargaSaham']
```

```python
MeanAvgTone
```

```python
DataSiapOlah=pd.DataFrame({'MeanAvgTone':MeanAvgTone,'MeanGoldsteinScale':MeanGoldsteinScale,'MeanNumMentions':MeanNumMentions,'SumAvgTone':SumAvgTone,'SumGoldsteinScale':SumGoldsteinScale,'SumNumMentions':SumNumMentions,'CountNews':CountNews,'NilaiSaham':NilaiSaham,'VolumeSaham':VolumeSaham,'IndexHargaSaham':IndexHargaSaham})
```

```python
import statsmodels.api as sm
```

```python
X=DataSiapOlah[['MeanAvgTone','MeanGoldsteinScale','MeanNumMentions']]
```

```python
y=DataSiapOlah['IndexHargaSaham']
```

```python
X=sm.add_constant(X)
```

```python
est = sm.OLS(y, X).fit()
est.summary()
```

```python
from sklearn.metrics import mean_squared_error
y_predict=1794.7569-290.19*DataSiapOlah['MeanAvgTone']+159.8747*DataSiapOlah['MeanGoldsteinScale']+384.1286*DataSiapOlah['MeanNumMentions']
```

```python
mean_squared_error(y, y_predict)
```

```python
t1_plot=y.plot(kind='line', title='Regresi', legend=True, fontsize=12, figsize=(10,5), label='y-true')
t2_plot=y_predict.plot(kind='line', title='Regresi', legend=True, fontsize=12, figsize=(10,5), label='y-predict')
t1_plot.set_xlabel("TahunBulan", fontsize=12)
t1_plot.set_ylabel("Nilai", fontsize=12)
```

```python
X=DataSiapOlah[['MeanAvgTone','MeanGoldsteinScale','MeanNumMentions','CountNews']]
X=sm.add_constant(X)
est = sm.OLS(y, X).fit()
est.summary()
```
