# Deskripsi Data
Data Produksi Perikanan Tangkap Nasional berisi data mengenai hasil penangkapan ikan yang disajikan menurut provinsi, nama ikan, jenis perairan, Periode, Volume Penangkapan(dalam ton), dan harga jual per jenis ikan (dalam rupiah)

## Analisis Data Statistik

```python
import csv
with open('volumeproduksiperikanan.csv') as csvfile:
    thefile=csv.reader(csvfile, delimiter=",", quotechar="'")
```

```python
import pandas as pd
```

```python
produksi_perikanan=pd.read_csv('volumeproduksiperikanan.csv', delimiter=',', encoding="ISO-8859-1")
```

```python
produksi_perikanan['NilaiJual'] = produksi_perikanan['Volume'] * produksi_perikanan['Nilai']
```

```python
produksi_perikanan.to_csv('new_ikan.csv')
```

```python
produksi_perikanan
```

```python
olah=produksi_perikanan[['NamaProvinsi','NamaIkan','Tahun','Volume','Nilai','NilaiJual']]
```

```python
olah
```

```python
olah=olah[olah.Volume.notnull()]
olah=olah[olah.Volume != 0]
```

```python
olah1 = olah.groupby(['NamaIkan']).agg({'Volume':sum})
```

```python
olah_urut=olah1.sort_values(['Volume'], ascending=False)
```

```python
olah_urut
```

```python
olah1_1=olah_urut.iloc[2:11]
```

```python
olah1_1
```

### Data di atas merupakan 10 data mengenai jenis ikan mana yang memiliki volume penangkapan terbanyak dari tahun 2012 hingga 2012

```python
olah1_plot = olah1_1.plot(kind='bar', title ="Jenis Ikan dengan Produksi Terbanyak", legend=True, fontsize=12)
olah1_plot.set_xlabel("Jenis Ikan", fontsize=12)
olah1_plot.set_ylabel("Total Produksi Ikan", fontsize=12)
```

```python
olah2_1=olah[olah.NamaIkan != "Ikan lainnya"]
```
